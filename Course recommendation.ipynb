```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import neattext.functions as nfx
```

```python
dataset=pd.read_csv("./udemy_course_data.csv")
dataset.head()
```

```python
dataset["course_title"]=dataset["course_title"].apply(nfx.remove_shortwords)
dataset["course_title"]=dataset["course_title"].apply(nfx.remove_special_characters)
```

```python
dataset["course_title"].head()
```

```python
countVector=CountVectorizer()
cv_mat=countVector.fit_transform(dataset["course_title"])
```

```python
vector_df=pd.DataFrame(cv_mat.todense(),columns=countVector.get_feature_names())
```

```python
vector_df
```

```python
cos_sim=cosine_similarity(vector_df)
```

```python
cos_sim.shape
```

```python
# drop duplicates 
course_index = pd.Series(dataset.index,index = dataset['course_title']).drop_duplicates()
course_index
```

```python
index=course_index['Financial Modeling Business Analysts Consultants']
index
```

```python
cos_values=list(enumerate(cos_sim[index]))
# cos_values=[value[1] for value in cos_values]
```

```python
indexes=[]
for (index,value) in cos_values:
    if value>0.5:
        indexes.append(index)
    
```
